```python
import os.path
from os import path
import pandas as pd
import tkinter as tk
import plotly.io as pio
from functools import partial
from PIL import ImageTk, Image
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from datetime import datetime, date, timedelta
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

%load_ext autoreload
%autoreload 1
%aimport BACKEND
```

```python
HEIGHT = 650
WIDTH = 1000
```

```python
root = tk.Tk()
root.title("STOCK ANALYSIS")
current_stock = ''

rootimageLabel = tk.Label(root)
rootimageLabel.pack()

background = ImageTk.PhotoImage(file = "images/background.png")
rootimageLabel.configure(image = background)
rootimageLabel.image = background

canvas = tk.Canvas(root, height = HEIGHT, width = WIDTH, bg = '#ffffff')
canvas.pack()

stock_frame = tk.Frame(root, bg = '#ffffff')#, bg = '#ffdbfb'
stock_frame.place(relx = 0.05, rely = 0.005, relwidth = 0.25, relheight = 0.9)

graph_frame = tk.Frame(root, bg = '#ffffff')#, bg = '#cfdafc'
graph_frame.place(relx = 0.35, rely = 0.15, relwidth = 0.6, relheight = 0.7)

time_frame = tk.Frame(graph_frame, bg = '#ffffff')#, bg = '#d6fccf'
time_frame.place(relx = 0.05, rely = 0.85, relwidth = 0.9 , relheight = 0.1)

plot_frame = tk.Frame(graph_frame, bg = '#ffffff')#, bg = '#d6fccf'
plot_frame.place(relx = 0, rely = 0.05, relwidth = 1 , relheight = 0.7)

imageLabel = tk.Label(plot_frame)
imageLabel.pack()

#, command = edit_text(j, '1 H')
def create_stock_buttons(stocks):
    for i, j in zip(range(1, len(stocks)+1), stocks):
        button = tk.Button(stock_frame, bd = 1, relief = 'ridge', text = "%s"%(j), command = partial(plot, '15 min', j))
        button.place(relx = 0.2, rely = 0.2*i, relwidth = 0.6, relheight = 0.1)


def create_time_buttons(stock, time_stamps = ['15 min', '1 H', '1 D', '1 W']):
    for i, j in zip(range(0, len(time_stamps)), time_stamps):
        button = tk.Button(time_frame, bd = 1, relief = 'ridge',text = "%s"%(j), command = partial(plot, j, stock))
        button.place(relx = 0.06*i*1.7, rely = 0.15, relwidth = 0.09, relheight = 0.7)
        
        
def create_TA_dropdown(technical_analysis_tools = ['SMA', 'EMA', 'RSI', 'Bollinger']):
    def_value = tk.StringVar(time_frame)
    def_value.set("-Technical Analysis Tools-")
    drop = tk.OptionMenu( time_frame, def_value, *technical_analysis_tools)
    drop.place(relx = 0.45, rely = 0.1, relwidth = 0.5 , relheight = 0.8)


def graph_visualisation(x_axis, y_axis, custom, label, duration):
    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x=x_axis,
            y=y_axis,
            customdata=custom,
            hovertemplate='<br>'.join([
                'Open : $%{y:.2f}',
                'Date : %{x}',
                '%{customdata[0]}',
                '%{customdata[1]}',
                '%{customdata[2]}'
            ]),
        )
    )

    fig.update_layout(xaxis_title="Date", yaxis_title=f'{label}')
#     fig.show()
    
    file_name = "images/{0}{1}.png".format(label, duration)
    if not path.exists(file_name):
        fig.write_image(file_name)
    
    image = ImageTk.PhotoImage(file = file_name)
    imageLabel.configure(image = image)
    imageLabel.image = image

def get_current_stock():
    return current_stock

def set_current_stock(stock):
    global current_stock
    current_stock = stock
    return current_stock
    
def plot(frequency, stock_name = get_current_stock()):
    datetime, close_val, extras = [], [], []
    if frequency == '1 W':
        data = BACKEND.weekly_data

    if frequency == '1 D':
        data =  BACKEND.daily_data

    if frequency == '1 H':
        data =  BACKEND.hourly_data
        stock_name = set_current_stock(stock_name)

    if frequency == '15 min':
        data =  BACKEND.intraday_data

    for i in data[stock_name]:
        datetime.append(i)
        close_val.append(float(data[stock_name][i]['Close']))
        extras.append(["Open: " + str(data[stock_name][i]['Open']), "High: " + str(data[stock_name][i]['High']), "Low: " + str(data[stock_name][i]['Low'])])

    extras.reverse()
    
    
    create_time_buttons(stock_name)
    create_TA_dropdown()
    graph_visualisation(datetime, close_val, extras, stock_name.upper(), frequency)
    

create_stock_buttons(BACKEND.stocks)
#create_time_buttons('GOOGL')

root.mainloop()
```
